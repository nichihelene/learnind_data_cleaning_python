```python
# let us remove from our file all the empty columns and rows , those who are not needed and reset the file 

file = file.dropna(how='all',axis = 0)

file = file.dropna(how='all',axis = 1)

file = file.reset_index(drop=True)

file.head(13)
```

```python
# let us create columns based on our desired output

columns = []

for col in file.columns:
    
    columns.append([str(s) for s in file[col][:1] if pd.notnull(s)])
    
print(columns)
```

```python
# I am renaming my columns 

columns = ['HDI_rank', 'countries','reference1','GEM_rank','GEM_value','reference_2','seats_parlia_wom','reference3','Fem_le_snr_offi_mana','reference_4','Fem_profes_techn_work','reference_5','Ratio_esti_fem_male_earn_incom','reference_6']


# let us reset the dataframe columns
file.columns = columns


# remove all rows(axis=0) that contains the informations not needed and let us reset the index

file = file.drop(np.arange(2),axis=0).reset_index(drop=True)

file.head()
```

```python
file.tail(120)
```

```python
# let us drop the rows from the bottom of the data we want 

file = file.drop(np.arange(180,len(file)), axis=0).reset_index(drop=True)

```

```python
file.tail(90)
```

```python
#let us add remove any row containing the HDI_level list element and put them in a column 

HDI_level_list = ["HIGH HUMAN DEVELOPMENT","MEDIUM HUMAN DEVELOPMENT","LOW HUMAN DEVELOPMENT"]



file = file.drop(file[file["HDI_rank"].isin(HDI_level_list)].index, axis=0).reset_index(drop=True)


#let us add a column to our data for the HDI level

file.loc[:, "HDI_level"] = np.where(file["HDI_rank"] > 155,"LOW HUMAN DEVELOPMENT",np.where(file["HDI_rank"] > 70,"MEDIUM HUMAN DEVELOPMENT","HIGH HUMAN DEVELOPMENT"))


#let us sort our data

file.sort_values(by="HDI_rank", inplace=True)


#let us display the head to see if the column was added

file.head()



```

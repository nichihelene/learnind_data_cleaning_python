```python
df = df.dropna(how="all",axis=1)
df = df.dropna(how="all",axis=0)
df = df.reset_index(drop=True)
df.head(6)
```

```python

columns =[]
for c in df.columns:
    columns.append([str(v) for v in df[c][0:1] if pd.notnull(v)])

print(columns)
    
```

```python
columns = ['HDI_rank','country','reference_1',
       'GEM_rank', 'GEM_value',
       'reference_2', 'Seats_parlia_wom', 'reference_3',
       'Fem _le_snr_offi_mana', 'reference_4',
       'Fem_profes_techn_work', 'reference_5',
       'Ratio_esti_ fem_male_earn_ incom','reference_6']

df.columns = columns



df = df.drop(np.arange(1), axis=0).reset_index(drop=True)
df.head()
```

```python
df = df.drop(np.arange(180,len(df)), axis=0).reset_index(drop=True)
df.tail(30)
```

```python

hdi_rank_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df = df.drop(df[df["HDI_rank"].isin(hdi_rank_categories)].index, axis=0).reset_index(drop=True)
df.loc[:, "HDI category"] = np.where(df["HDI_rank"] > 155,"LOW HUMAN DEVELOPMENT",np.where(df["HDI_rank"] > 70,"MEDIUM HUMAN DEVELOPMENT","HIGH HUMAN DEVELOPMENT"))
df.sort_values(by="HDI_rank", inplace=True)
df.head()

```

```python
import pandas as pd
import numpy as np
import whyqd as Wh


#let us define a structure for our data.

schema = Wh.Schema()

details = {
   "name": "Gender_Empowerment_Measure_Per_Country_Table_29_Nojh",
   "title": "Gender empowerment measure and achieving equality for all women and men ",
   "description": "Database that ranks countries based on the gender empowerment measure taken by them  in achieving "
                  "equality for all women and men ",
}
schema = Wh.Schema()
schema.set_details(**details)




```

```python
# let us define our field in our schema

fields = [
      {
        "name": "Indicator_name",
        "title":"Indicator Name",
        "type": "string",
        "description": "Indicator described in the data series.",
        "constraints": {
          "minimum": 1
        }
      },
      {
        "name": "Hum_dev_level",
        "title": "Human Development Level(HDL)",
        "type": "string",
        "description": "Human Development Level of all the countries from those who have the highest human development "
                       "level to those who have  the lowest human development and it is determined on the basis of "
        
        
      },
      {
        "name": "Country_name",
        "title": "Country Name",
        "type": "string",
        "description": "Name  of the countries",
        "constraints": {
          "maximum": 4,
          "required": True
        }
      },
      
      {
        "name": "Reference ",
        "title": "Reference",
        "type": "string",
        "description": "Reference to data series",
        
        
      },
      {
        "name": "values ",
        "title": "values",
        "type": "number",
        "description": "value for the indicator Name",
        
        
      },
      
 
    ]


#if __name__ == "__main__":

for field in fields:
        schema.set_field(**field)

print(schema.default_field_settings("string"))
```

```python
# I have created A DIRECTORY and a FILENAME where my schema will be saved

directory = "Jasminehelene"
filename = "new_json_schema"
schema.save(directory, filename,overwrite=True )
```

```python
# let us import a new librairies called pandas_schema that do offers items to validate our dataframe


from pandas_schema import Column, Schema
from pandas_schema.validation import LeadingWhitespaceValidation, TrailingWhitespaceValidation, IsDtypeValidation, InListValidation

# let us test some columns
columns = ["country_name", "hum_dev_level", "values_"]
# let us test our categories
HDI_level_list = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]

schema = Schema([
    Column("country_name", [LeadingWhitespaceValidation(), TrailingWhitespaceValidation()]),
    Column("hum_dev_level", [InListValidation(HDI_level_list)]),
    Column("values_", [IsDtypeValidation(np.dtype(float)), IsDtypeValidation(np.dtype(int))])
])

errors = schema.validate(df[columns])

print(F"Nombre d'erreurs :  {len(errors)}")
# Let us if we have some errors
for error in errors[:]:
    print(error)
```
